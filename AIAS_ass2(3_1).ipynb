```python
#Question 1: Zero-Shot Prompting
def is_palindrome(num):
    original = num
    reverse = 0

    while num > 0:
        digit = num % 10
        reverse = reverse * 10 + digit
        num //= 10

    return original == reverse


n = int(input("Enter a number: "))
if is_palindrome(n):
    print("Palindrome Number")
else:
    print("Not a Palindrome Number")
```

```python
#Question 2: One-Shot Prompting
def factorial(n):
    if n < 0:
        return "Factorial not defined for negative numbers"

    result = 1
    for i in range(1, n + 1):
        result *= i

    return result


n = int(input("Enter a number: "))
print("Factorial:", factorial(n))
```

```python
#Question 3: Few-Shot Prompting
def is_armstrong(num):
    digits = len(str(num))
    total = 0
    temp = num

    while temp > 0:
        digit = temp % 10
        total += digit ** digits
        temp //= 10

    return total == num

n = int(input("Enter a number: "))
if is_armstrong(n):
    print("Armstrong Number")
else:
    print("Not an Armstrong Number")
```

```python
#Question 4: Context-Managed Prompting
def classify_number(n):
    if n <= 1:
        return "Neither Prime nor Composite"

    for i in range(2, int(n ** 0.5) + 1):
        if n % i == 0:
            return "Composite Number"

    return "Prime Number"


n = int(input("Enter a number: "))
print(classify_number(n))
```

```python
#Question 5: Zero-Shot Prompting
def is_perfect(num):
    if num <= 0:
        return False

    total = 0
    for i in range(1, num):
        if num % i == 0:
            total += i

    return total == num


n = int(input("Enter a number: "))
if is_perfect(n):
    print("Perfect Number")
else:
    print("Not a Perfect Number")
```

```python
#Question 6: Few-Shot Prompting
try:
    n = int(input("Enter a number: "))

    if n % 2 == 0:
        print("Even")
    else:
        print("Odd")

except ValueError:
    print("Invalid input. Please enter an integer.")
```
